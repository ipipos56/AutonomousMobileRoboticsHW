```python
# GitHubLink 
# https://github.com/ipipos56/AutonomousMobileRoboticsHW

import matplotlib.pyplot as plt
import numpy as np
import sympy
from sympy import symbols, integrate
import json


info = json.load(open('info.json'))

t = []
t_cur = 0
Ts = 0.03
for i in range(max(len(info['all_q']), len(info['all_rq']))):
  t.append(t_cur)
  t_cur += Ts

all_q = np.array([info['all_q']]).squeeze()
all_rq = np.array([info['all_rq']]).squeeze()
control_points=np.array([[3,0], [6,4], [3,4], [3,1], [0,3]])


# green - calculated by integration, red - ROS Odometry, blue - input points
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(all_q[:, 0], all_q[:, 1], color='green')
ax.plot(all_rq[:, 0], all_rq[:, 1], color='red')
ax.plot(control_points[:, 0], control_points[:, 1], color='blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
